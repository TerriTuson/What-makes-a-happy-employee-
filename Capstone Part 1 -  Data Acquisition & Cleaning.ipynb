```python
#convert Lithuanian salaries from Lev to Euros
salary["Q50_23_1"] = 0.5113076272 * salary["Q50_23_1"]
```

```python
#convert Croatian salaries from HRK to Euros
salary["Q50_25_1"] = 0.1311363250 * salary["Q50_25_1"]
```
